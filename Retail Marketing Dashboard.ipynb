```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month']=df['Date'].dt.month
monthly_sales = df.groupby('Month')['Total_Sales'].sum()
```

#DATA VISUALIZATION

```python
plt.figure(figsize=(20,10))
plt.suptitle("RETAIL SALES DASHBOARD",fontsize=18)
plt.subplot(2,3,1)
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts.values,labels=gender_counts.index,autopct='%1.1f%%',startangle=90)
plt.title("Customer Gender Distribution")
plt.subplot(2,3,2)
sns.barplot(x="Product Category",y="Total_Sales",data=df,estimator=sum,color='purple')
plt.title("Total Sales by Category")
plt.subplot(2,3,3)
sns.barplot(x="Product Category",y="Quantity",data=df,estimator=sum,color='yellow')
plt.title("Quantity Sold by Category")
plt.subplot(2,3,4)
sns.boxplot(x=df["Price per Unit"],color='red')
plt.title("Product Price Distribution")
plt.subplot(2,3,5)
sns.barplot(x="Gender",y="Total_Sales",data=df,estimator=sum)
plt.title("Total Sales by Gender")
plt.subplot(2,3,6)
plt.plot(monthly_sales.cumsum())
plt.xlabel("Month")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Sales Growth")

plt.tight_layout()
plt.show()
```
